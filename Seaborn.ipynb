# Introduction

`seaborn` is a Python data visualization library, an alternative to `matplotlib` (actually, it is built upon `matplotlib`). It is useful for advanced visualization graphics and informative statistical graphs.
For some introductory notes on seaborn, refer to this documentaion: https://seaborn.pydata.org/tutorial/introduction.html

Let's begin

```python
# import the package
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Load data

```python
# import the built-in dataset `tips` in seaborn
tips = sns.load_dataset("tips")
tips.info()
```

```python
tips.head()
```

# Relational Plots

```python
sns.relplot(data= tips, x = 'total_bill', y = 'tip')
```

```python
#categorize the tips by sex in the same plot
sns.relplot(data=tips, x="total_bill", y="tip", hue = "sex") #categorize the tips by sex
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue = "day") #categorize the tips by day
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue = "time") #categorize the tips by lunch/dinner time
```

```python
#categorize the tips further by time and view them side by side
sns.relplot(data=tips, x="total_bill", y="tip", hue = "smoker", col='time') #categorize the tips by smoker/non-smoker and further by time
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue = "size") #categorize the tips by size like hotmaps
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue = "size", size="size")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue = "size", size="size", sizes = (15,200)) #to increase the data point size and make it more informative
```

```python
# to remove overlapping of data point and specify transparency, use `alpha`
sns.relplot(data=tips, x="total_bill", y="tip", hue = "size", size="size", sizes = (15,200), alpha=0.5)
```

```python
sns.relplot(data=tips, x='day', y='total_bill')
```

```python
#to show the relation using line. dark line represents the mean and light blue area represents distribution of data
sns.relplot(data=tips, x='day', y='total_bill', kind='line')
```

```python
# now use `hue` to see the trend based upon gender
sns.relplot(data=tips, x='day', y='total_bill', kind='line', hue='sex')
```

```python
# now divide this further using `col` to see the trend based on time
sns.relplot(data=tips, x='day', y='total_bill', kind='line', hue='sex', col='time')
```

# Distribution Plots

```python
sns.displot(data=tips, x='total_bill') #similar to histograms of matplotlib
```

```python
#let's see how this distribution changes based on the customer's gender side by side using `col`
sns.displot(data=tips, x='total_bill', col='sex')
```

```python
#let's see how this distribution changes based on the customer's gender in the same plot using `hue`
sns.displot(data=tips, x='total_bill', hue='sex')
#overlapped part is in different color
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', col='day')
```

```python
# now to see how the variables are distributed as continuous variables
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde')
```

```python
# to draw stacked distributions
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', multiple='stack')
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde', multiple='stack', col='day')
```

```python
# using multiple attributes
sns.displot(data=tips, x='total_bill', y='tip', kind='kde')
# this distribution can be viewed like centre as tip of mountain and foot of the mountain is along the outercircles.
# it is like a 2-d projection of a 3-d mountain on a plane of paper
```

```python
#distinguish by gender
sns.displot(data=tips, x='total_bill', y='tip', kind='kde', hue='sex')
```

```python
#to view the centerpoint of distribution by looking at datapoints projected on x and y axes
sns.displot(data=tips, x='total_bill', y='tip', kind='kde', hue='sex', rug=True)
```

```python
#looking at the cumulative distribution function
sns.displot(data=tips, x='total_bill', kind='ecdf')
```

```python
sns.displot(data=tips, x='total_bill', kind='ecdf', hue='sex')
```

```python
sns.displot(data=tips, x='total_bill', kind='ecdf', hue='sex', col='day')
```

# Categorical Plots
Useful for variables that have discrete values/limited number of values likes days in a week

```python
sns.catplot(data=tips, x='day', y='total_bill')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', hue='sex')
```

```python
#boxplot with median as centre, first and third quartile as two ends of the box
# the bars have the theoretical minimum and maximum (calculated errors), not actual min. or max.
# outliers are marked as data points outside the box/error limits
sns.catplot(data=tips, x='day', y='total_bill', kind='box')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='box', hue='sex')
```

```python
# violin plot - another type of box, just made as violin (sides are like distribution of say total bill for each day)
# dots are version of box, representing median, quartile 1 and 3 and then the theoretical errors are shown by the line
# everything outside are outliers
sns.catplot(data=tips, x='day', y='total_bill', kind='violin')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='violin', hue='sex')
```

```python
#to view both genders on same plot
sns.catplot(data=tips, x='day', y='total_bill', kind='violin', hue='sex', split=True)
```

```python
#if you don't want to specify the kind of plot, you can call the violin plot directly
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True, inner='quartile')
```

```python
# bar plot with top of bar as mean(or is it median?), the error bar shows Q1 and Q3
sns.catplot(data=tips, x='day', y='total_bill', kind='bar')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='bar', hue='sex')
```

# Joint Plots

```python
# showing the distribution of 2 attributes together as well as for each attribute, all in one plot
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='kde')
```

```python
kind = ['scatter', 'hist', 'hex', 'kde', 'reg', 'resid']
for k in kind:
    sns.jointplot(data=tips, x='total_bill', y='tip', kind=k)
```

# Pair plot
Really useful when you don't have time to plot relational plot for all
numerical variables and want to get a big picture of the dataset by looking at all pairwise plots.
Remember: This doesn't look at categorical variables for e.g. Rating of service that has values 'Good', 'Bad', 'Need to Improve'

However, if you have 100 numerical variables, then it will generate a 100 x 100 grid of these plots, which is time consuming.

```python
sns.pairplot(data=tips)
```

```python
sns.pairplot(data=tips, hue='sex')
```
